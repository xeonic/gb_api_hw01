1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json

```python
import requests
import json
```

```python
url = 'https://api.github.com/users/xeonic/repos'
req = requests.get(url)
req.status_code
```

```python
if req.status_code == 200:
    with open('xeonic.json', 'w') as json_file:
        json.dump(req.json(), json_file)
else:
    print('Error')
```

2. Изучить список открытых API (https://www.programmableweb.com/category/all/apis). Найти среди них любое, требующее авторизацию (любого типа). Выполнить запросы к нему, пройдя авторизацию. Ответ сервера записать в файл.

```python
bearer_token = 'secret BEARER_TOKEN :)'
```

```python
search_url = "https://api.twitter.com/2/tweets/search/recent"
query_params = {'query': 'from:P_Fiala',
                'tweet.fields': 'created_at', 
                'expansions': 'author_id', 
                'user.fields': 'created_at'}
```

```python
def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """

    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2RecentSearchPython"
    return r
```

```python
def connect_to_endpoint(url, params):
    response = requests.get(url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
req = connect_to_endpoint(search_url, query_params)

with open('fiala.json', 'w') as json_file:
        json.dump(req, json_file)
```
